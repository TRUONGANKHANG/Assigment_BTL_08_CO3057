```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


bg = cv2.imread("input/target_3.jpg")
src = cv2.imread("input/source_3.jpg")
mask = cv2.imread("input/mask_3.bmp", 0)

bg = cv2.cvtColor(bg, cv2.COLOR_BGR2RGB)
src = cv2.cvtColor(src, cv2.COLOR_BGR2RGB)

h, w = bg.shape[:2]
src = cv2.resize(src, (w, h))
mask = cv2.resize(mask, (w, h))

mask_bin = (mask > 127).astype(np.uint8)


blended_copy = bg.copy()
blended_copy[mask_bin == 1] = src[mask_bin == 1]


plt.figure(figsize=(6, 5))
plt.imshow(blended_copy)
plt.title("Direct Copy-Paste Result")
plt.axis("off")
plt.show()
```
